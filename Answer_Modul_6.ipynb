```
img1 = Image.open('f4.jpg')
img2 = Image.open('f9.jpg')
img3 = Image.open('f11.jpg')
img4 = Image.open('f15.jpg')

img360 = img4.rotate(360)

r90 = img1.transpose(Image.ROTATE_90)
r180 = img2.transpose(Image.ROTATE_180)
r270 = img3.transpose(Image.ROTATE_270)
r360 = img360.transpose(Image.FLIP_LEFT_RIGHT)

n5 = r90.resize((256,256))
n6 = r180.resize((256,256))
n7 = r270.resize((256,256))
n8 = r360.resize((256,256))

nr_imgf = Image.new('RGB', (512,512), (250,250,250))

nr_imgf.paste(n5,(0,0))
nr_imgf.paste(n6,(256,0))
nr_imgf.paste(n7,(0,256))
nr_imgf.paste(n8,(256,256))

nr_imgf.save('rf.jpg', 'JPEG')
```

```
Img5 = Image.open('rf.jpg')
Img6 = Img5.transpose(Image.FLIP_LEFT_RIGHT)

nr_imgf = Image.new('RGB', (1024,512), (250,250,250))
nr_imgf.paste(Img5,(0,0))
nr_imgf.paste(Img6,(512,0))

nr_imgf.save('rfnew.jpg', 'JPEG')
nr_imgf
```

```
nrf = Image.open('rfnew.jpg')
print('Format type : '+str(nrf.format))
print('Size : '+str(nrf.size[0])+' x '+str(nrf.size[1]))
```

```
import numpy as np
import imageio
import matplotlib.pyplot as plt
pic = imageio.imread('f4.jpg')

fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), ax):
    
    # create zero matrix
    split_img = np.zeros(pic.shape, dtype="uint8") # 'dtype' by default: 'numpy.float64'
    
    # assing each channel 
    split_img[ :, :, c] = pic[ :, :, c]
    
    # display each channel
    ax.imshow(split_img)
```
